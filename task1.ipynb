```python
import numpy as np
file_object = open("./glove.6B.50d.txt", "r")

# function that computes cosine similarity between 2 vectors
def orientation(first_array, second_array):
    
    dp = np.dot(first_array, second_array)
    first_array_norm = np.linalg.norm(first_array)
    second_array_norm = np.linalg.norm(second_array)
    
    return dp / (first_array_norm * second_array_norm)

elbert = False
lines = file_object.read().splitlines()
elbert_vector = []

for line in lines:
    tokens = line.split()
    token = tokens[0]
    if token == "elbert":
        elbert = True
        elbert_vector = tokens

        
print("Does my name exist in the dataset? " + str(elbert) + "\n")
print("Vector representation of it:\n")
lk = elbert_vector[1:]

# typecast values into floats
lb = [float(i) for i in lk]

# print out the 50 dimensional vector
print(lb)
print("\n")

# now, compare the elbert vector with all of the other vectors (besides the elbert vector)
# and store 5 vectors with the highest cosine similarity 

first = 0
first_word = None
second = 0
second_word = None
third = 0
third_word = None
fourth = 0
fourth_word = None
fifth = 0
fifth_word = None

for line in lines:
    tokens = line.split()
    token = tokens[0]
    
    # if the word is elbert, skip
    if token == "elbert":
        continue
        
    vector = tokens[1:]
    vector = [float(i) for i in vector]
    
    # cs is cosine similarity
    cs = orientation(lb, vector)
    
    if cs > first:
        first = cs
        first_word = token
    elif cs > second:
        second = cs
        second_word = token
    elif cs > third:
        third = cs
        third_word = token
    elif cs > fourth:
        fourth = cs
        fourth_word = token
    elif cs > fifth:
        fifth = cs
        fifth_word = token
    else:
        continue
    
top_5_words = []
top_5_words.append(first_word)
top_5_words.append(second_word)
top_5_words.append(third_word)
top_5_words.append(fourth_word)
top_5_words.append(fifth_word)

top_5_cs = []
top_5_cs.append(first)
top_5_cs.append(second)
top_5_cs.append(third)
top_5_cs.append(fourth)
top_5_cs.append(fifth)


print(top_5_words)
print(top_5_cs)

# print("Top 5: \n")
# print(top_5)
# print(top_5_words)
            
file_object.close()
```
