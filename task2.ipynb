```python
import numpy as np
from __future__ import division

file_object = open("./glove.6B.50d.txt", "r")

# function that computes cosine similarity between 2 vectors
def orientation(first_array, second_array):
    
    dp = np.dot(first_array, second_array)
    first_array_norm = np.linalg.norm(first_array)
    second_array_norm = np.linalg.norm(second_array)
    
    return dp / (first_array_norm * second_array_norm)

# save the lines into a variable
lines = file_object.read().splitlines()

# lowercased string
string = "this is a sexy provocative sentence"
string = string.lower()
string = string.split()

counter = 0
words = []
cs = []

while counter < len(string):
    word = string[counter]
    
    for line in lines:
        tokens = line.split()
        token = tokens[0]
        if word == token:
            words.append(word)
            pre_cs = tokens[1:]
            pre_cs = [float(i) for i in pre_cs]
            cs.append(pre_cs)
    
    counter = counter + 1

# component-wise sum of vectors
vector = [sum(x) for x in zip(*cs)]
final_vector = [x / len(cs) for x in vector]

# top 5
first = 0
first_word = None
second = 0
second_word = None
third = 0
third_word = None
fourth = 0
fourth_word = None
fifth = 0
fifth_word = None

# check for the top similar vector
for line in lines:
    tokens = line.split()
    token = tokens[0]

    v1 = tokens[1:]
    v2 = [float(i) for i in v1]
    
    highest_cs = orientation(final_vector, v2)

    if highest_cs > first:
        first = highest_cs
        first_word = token
    elif highest_cs > second:
        second = highest_cs
        second_word = token
    elif highest_cs > third:
        third = highest_cs
        third_word = token
    elif highest_cs > fourth:
        fourth = highest_cs
        fourth_word = token
    elif highest_cs > fifth:
        fifth = highest_cs
        fifth_word = token
    else:
        continue
    
top_5_words = []
top_5_words.append(first_word)
top_5_words.append(second_word)
top_5_words.append(third_word)
top_5_words.append(fourth_word)
top_5_words.append(fifth_word)

top_5_cs = []
top_5_cs.append(first)
top_5_cs.append(second)
top_5_cs.append(third)
top_5_cs.append(fourth)
top_5_cs.append(fifth)

print(top_5_words)
print(top_5_cs)

file_object.close()
```
