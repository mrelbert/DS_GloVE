```python
import numpy as np
from __future__ import division

file_object = open("./glove.6B.50d.txt", "r")

# function that computes cosine similarity between 2 vectors
def orientation(first_array, second_array):
    
    dp = np.dot(first_array, second_array)
    first_array_norm = np.linalg.norm(first_array)
    second_array_norm = np.linalg.norm(second_array)
    
    return dp / (first_array_norm * second_array_norm)

# save the lines into a variable
lines = file_object.read().splitlines()

# lowercased string
s0 = "nissan and audi and bmw and hyundai are car brands"
s0 = s0.lower()
s0 = s0.split()

c0 = 0
w0 = []
cs0 = []

while c0 < len(s0):
    word = s0[c0]
    
    for line in lines:
        tokens = line.split()
        token = tokens[0]
        if word == token:
            w0.append(word)
            pre_cs = tokens[1:]
            pre_cs = [float(i) for i in pre_cs]
            cs0.append(pre_cs)
    
    c0 = c0 + 1

# component-wise sum of vectors
v0 = [sum(x) for x in zip(*cs0)]
final_v0 = [x / len(cs0) for x in v0]

s1 = "mercedes and ford and honda are car brands as well"
s1 = s1.lower()
s1 = s1.split()

c1 = 0
w1 = []
cs1 = []

while c1 < len(s1):
    word = s1[c1]
    
    for line in lines:
        tokens = line.split()
        token = tokens[0]
        if word == token:
            w1.append(word)
            pre_cs = tokens[1:]
            pre_cs = [float(i) for i in pre_cs]
            cs1.append(pre_cs)
    
    c1 = c1 + 1

# component-wise sum of vectors
v1 = [sum(x) for x in zip(*cs1)]
final_v1 = [x / len(cs1) for x in v1]

s2 = "hydrogen and oxygen and helium are chemical elements"
s2 = s2.lower()
s2 = s2.split()

c2 = 0
w2 = []
cs2 = []

while c2 < len(s2):
    word = s2[c2]
    
    for line in lines:
        tokens = line.split()
        token = tokens[0]
        if word == token:
            w2.append(word)
            pre_cs = tokens[1:]
            pre_cs = [float(i) for i in pre_cs]
            cs2.append(pre_cs)
    
    c2 = c2 + 1

# component-wise sum of vectors
v2 = [sum(x) for x in zip(*cs2)]
final_v2 = [x / len(cs2) for x in v2]

# cosine similarity of s0 and s1
s0_and_s1 = orientation(final_v0, final_v1)

# cosine similarity of s0 and s2
s0_and_s2 = orientation(final_v0, final_v2)

print("Cosine similarity between s0 and s1: " + str(s0_and_s1))
print("Cosine similarity between s0 and s2: " + str(s0_and_s2))
```
